# Quick Demo
This notebook demonstrates synthesizing a signal and running pipelines.

```python
from pwvstudy.synthetic import synthesize_ppg_bundle
from pwvstudy.pipelines import SegmentalProcessing, TransitTimeMethod, CompositePWV
import pandas as pd
b = synthesize_ppg_bundle(seconds=10, fs=1000, hr=70, seed=1)
prox = b['ppg']; dist = prox.copy(); import numpy as np; dist = np.roll(dist, int(0.08*1000))
seg = SegmentalProcessing(fs=1000)
ttm = TransitTimeMethod(fs=1000)
cmp = CompositePWV(fs=1000)
r1, r2, r3 = seg.run(prox, dist, 0.8), ttm.run(prox, dist, 0.8), cmp.run(prox, dist, 0.8)
pd.DataFrame([r1,r2,r3], index=['Segmental','TT','Composite'])
```
