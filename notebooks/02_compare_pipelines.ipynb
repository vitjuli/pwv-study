# Compare Pipelines under Noise/Artifacts
Parameter sweep on synthetic data (SNR, baseline drift, motion steps).

```python
import numpy as np, pandas as pd
from pwvstudy.synthetic import synthesize_ppg_bundle
from pwvstudy.pipelines import SegmentalProcessing, TransitTimeMethod, CompositePWV
fs=1000; seg=SegmentalProcessing(fs); tt=TransitTimeMethod(fs); cmp=CompositePWV(fs)
rows = []
for snr in [10,15,20]:
    for steps in [0,2,4]:
        b = synthesize_ppg_bundle(seconds=12, fs=fs, hr=70, seed=snr+steps,
                                 noise_snr_db=snr, motion_steps=steps)
        prox = b['ppg']; dist = np.roll(prox, int(0.08*fs))
        for name, pipe in [('Segmental',seg),('TT',tt),('Composite',cmp)]:
            r = pipe.run(prox, dist, 0.8)
            r.update({'snr':snr,'steps':steps,'pipeline':name})
            rows.append(r)
pd.DataFrame(rows)
```
